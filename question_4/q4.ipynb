```python
import requests
import pandas as pd
import ast
import matplotlib.pyplot as plt
```

```python
dates = list(pd.date_range(start='2020-03-01', periods=30))
num_confirmed_cases = []
for date in dates:
    date_str = date.strftime('%Y-%m-%d')
    api_str = f"https://covid-api.com/api/reports/total?date={date_str}&iso=SGP"
    resp = requests.get(api_str)
    
    num_confirmed_cases.append(ast.literal_eval(resp.content.decode('utf-8'))['data']['confirmed'])    

print(num_confirmed_cases)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(dates, num_confirmed_cases, marker='o', linestyle='-', color='b', label='COVID-19 Cases')

# graph labelling
plt.title('COVID-19 Case Growth Trend', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Cumulative Cases', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
